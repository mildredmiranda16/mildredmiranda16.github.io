## Proyecto propio: Mi horario

```python
# Llamar las paqueterías
import pandas as pd # Para manejar datos
import matplotlib.pyplot as plt  #Para hacer tablas
from matplotlib import colors # Para agreugar colores
```

```python
# Datos del horario (puedes personalizar las clases, días y horas)
data = {
    'Hora': ['7:00-9:00', '9:00-10:30', '10:30-12:30', '12:30-14:00', '14:00-16:00', '16:00-19:00'],
    'Lunes': ['Servicio', 'Servicio', 'Servicio','TS de Macro','',''],
    'Martes': ['Servicio', 'Estadística','Servicio', 'Python','', 'Democracia',],
    'Miércoles': ['Servicio', 'Servicio', 'Servicio','TS de Macro','',''],
    'Jueves': ['Servicio','Estadística', '', 'Python', '',''],
    'Viernes': ['Análisis IE','Análisis IE','','','','']
}
```

```python
# Crear data frame
df = pd.DataFrame(data)
```

```python
# Colores asignados a cada materia
color_dict = {
    'Servicio': '#87CEFA',        # Azul
    'TS de Macro': '#ff7f0e',
    'Estadística': '#9467bd',
    'Python': '#DDA0DD',       # Lila
    'Democracia': '#9ACD32',     # Verde
    'Análisis IE': '#FFFF00',     # Amarillo
    '': '#FFFFFF'              # Blanco para las celdas vacías
}
```

```python
# Crear la figura
fig, ax = plt.subplots(figsize=(8, 6))  # Ajusta el tamaño de la figura
# Eliminar el eje
ax.axis('tight')
ax.axis('off')
# Crear la tabla
tabla = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')
# Aplicar los colores a las celdas
for i in range(len(df)):
    for j in range(len(df.columns)):
        cell_value = df.iloc[i, j]
        if cell_value in color_dict: # Check if the cell value is a valid key in the color dictionary
            tabla[(i+1, j)].set_facecolor(color_dict[cell_value])

# Ajustar el tamaño de la fuente
tabla.auto_set_font_size(False)
tabla.set_fontsize(14)

# Ajustar el tamaño de las celdas
tabla.scale(1.5, 2.5)
# Mostrar la tabla
plt.show()
```
