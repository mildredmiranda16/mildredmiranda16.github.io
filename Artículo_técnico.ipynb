```python
# Llamar las librerias conocidas que utilizaremos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sb

%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
# Para el modelo predictivo utilizaremos la libreria scikit learn
# Así que cargaremos las librerias

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest
```

```python
#Leer el data set
basecar = pd.read_excel("/content/Mildred Miranda Anguas - insurance_challenge_train.xlsx", sheet_name= "Sheet1")
```

```python
# Modificar la columna APERSAUT: 0 se queda igual, del 1 al 7 se convierte en 1
basecar['APERSAUT'] = basecar['APERSAUT'].apply(lambda x: 0 if x == 0 else 1)

# Modificar la columna MZPART: 0 se queda igual, del 1 al 9 se convierte en 1
basecar['MZPART'] = basecar['MZPART'].apply(lambda x: 0 if x == 0 else 1)

# Modificar la columna MFWEKIND: 0 se queda igual, del 1 en adelante se convierte en 1
basecar['MFWEKIND'] = basecar['MFWEKIND'].apply(lambda x: 0 if x == 0 else 1)
```

```python
# Seleccionar solo las columnas especificadas
colsel = ['APERSAUT', 'MZPART', 'MFWEKIND']
tabla = basecar[colsel]

# Mostrar la tabla resultante
print(tabla.head(10)) #Ver las primeras 10 lineas de la nueva tabla
```

```python
# Verificar los cambios
print(tabla['APERSAUT'].value_counts())
print(tabla['MZPART'].value_counts())
print(basecar['MFWEKIND'].value_counts())
```

```python
#Divido las variables etiquetas
X=tabla.drop(['APERSAUT'], axis=1)
y=tabla['APERSAUT']# target o objetivo
```

```python
# Split dataset in training and test datasets
X_train, X_test = train_test_split(tabla, test_size=0.2, random_state=6)
y_train =X_train["APERSAUT"]
y_test = X_test["APERSAUT"]
```

```python
X_train.head()
```

```python
tablax = ['MZPART', 'MFWEKIND']
X_train[tablax].head()
```

```python
from sklearn.naive_bayes import GaussianNB

# objeto del modelo
gnb = GaussianNB()
# entranar el modelo
gnb.fit(X_train[tablax].values,y_train)

# Hacemos las prediciones
y_pred = gnb.predict(X_test[tablax])

print('Precisión en el set de Entrenamiento: {:.2f}'
     .format(gnb.score(X_train[tablax], y_train)))
print('Precisión en el set de Test: {:.2f}'
     .format(gnb.score(X_test[tablax], y_test)))
```
