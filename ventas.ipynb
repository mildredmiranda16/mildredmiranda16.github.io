```python
## Importar librerías

from sklearn.linear_model import LinearRegression as lm  ## Hacer modelos
import pandas as pd ## manejador de tablas o data frames, cargar csv
import matplotlib.pyplot as plt ## graficas.
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/model.csv'
df = pd.read_csv(path, sep=',')
df.head()
```

## Regresión lineal entre las ventas y el gasto en TV

```python
## Selección columnas de pandas
x = df['Sales'] # Valor de las ventas
y = df['TV'] # Valor de TV
```

```python
## Función scatter de matplotlib (plt)
plt.scatter (x, y)
#Agregar etiquetas
plt.xlabel('Ventas')
plt.ylabel('Gasto en Marketing de TV')
plt.title('Relación entre las ventas y el marketing en Televisión')
```

```python
x = df[['Sales']]  #construyendo df
y = df ['TV'] #construyendo serie o vector
```

```python
model = lm() # Asignar la funcion para hacer el modelo
# Calcula con el vector "y" y el dataframe por la correlación R
model.fit(x, y) ## fit significa ajuste
# Muestra cuál es la correlación entre x y y
model.score(x, y)
```

```python
y_fit = model.predict(x)
```

```python
# ahora todo junto
plt.scatter (x, y)
#Agregar etiquetas
plt.xlabel('Ventas')
plt.ylabel('Gasto en Marketing de TV')
plt.title('Relación entre las ventas y el gasto de Marketing en Televisión')

plt.plot(x, y_fit, color= 'red')
plt.show()
```
